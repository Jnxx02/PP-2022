1. Menampilkan data bertipe CSV

```python
import pandas as pd
data = pd.read_csv('Pelayanan Akademik.csv')
data.head()
```

2. Menampilkan data yang dibutuhkan untuk perhitungan CSI

```python
data = data.iloc[:,1:15]
data.head()
```

3. Konversi data ke numerik berdasarkan pilihannya

```python
# Dictionary konversi
convert = {'STB' : 1, 'TB' : 2, 'C' : 3, 'B' : 4, 'SB' : 5}
```

```python
# Program konversi
data = data.replace(convert)
data.head()
```

4. Menampilkan data opini/kenyataan saja

```python
# Data Opini = d_opini 
d_opini = data.iloc[:,[2*i for i in range(0,7)]]
d_opini.head()
```

5. Menampilkan data harapan saja

```python
# Data Harapan = d_harapan
d_harapan = data.iloc[:,[2*i+1 for i in range(0,7)]]
d_harapan.head()
```

6. Menampilkan tabel dataframe berdasarkan pertanyaan

```python
# Menghitung rataan opini/kenyataan pertanyaan (XP)
import numpy as np
XP = np.array(d_opini).mean(axis = 0)
```

```python
# Menghitung rataan harapan pertanyaan (YP)
YP = np.array(d_harapan).mean(axis = 0)
```

```python
# Menghitung bobot pertanyaan (WP) dinyatakan dalam persen(%)
WP = (np.array(YP) / np.array(YP).sum()) * 100
```

```python
# Menghitung kenyataan bobot pertanyaan (KBP)
KBP = np.array(WP) * np.array(XP)
```

```python
# Membuat dataframe pertanyaan (DFP)
data_pertanyaan = {'Kenyataan (X)' : [x for x in XP],
        'Harapan (Y)' : [y for y in YP],
        'Bobot (W)' : [w for w in WP],
        'W x X' : [k for k in KBP]}
idx = [i for i in range(1, len(XP) + 1)]
DFP = pd.DataFrame(data_pertanyaan, index = [idx])
DFP
```

7. Menampilkan kategori CSI pertanyaan

```python
# Menghitung CSI pertanyaan
CSI_P = np.array(KBP).sum() / 5
CSI_P.round(2)
```

```python
# Kategori nilai CSI
if CSI_P >= 80.01:
    Kriteria = 'CSI : Sangat Puas'
elif CSI_P >= 65.01 and CSIP <= 80.00:
    Kriteria = 'CSI : Puas'
elif CSI_P >= 50.01 and CSIP <= 65.00:
    Kriteria = 'CSI : Cukup Puas'
elif CSI_P >= 35.01 and CSIP <= 50.00:
    Kriteria = 'CSi : Kurang Puas'
else:
    Kriteria = 'CSI : Tidak Puas'
Kriteria
```

8. Membuat grafik IPA untuk pertanyaan

```python
from matplotlib import pyplot as plt
# Judul dan keterangan grafik
plt.title('Indicator Performance Analysis (IPA) Pertanyaan')
plt.xlabel('Kenyataan')
plt.ylabel('Harapan')

# Isi grafik (titik dan garis) serta keterangannya (legenda)
plt.scatter(XP, YP, color = 'grey', label = 'Data')
plt.plot([XP.mean(),XP.mean()],[YP.min()-0.1,YP.max()+0.1], color = 'red', label = 'Rataan Kenyataan')
plt.plot([XP.min()-0.2,XP.max()+0.2],[YP.mean(),YP.mean()], color = 'blue', label = 'Rataan Harapan')
plt.legend()
plt.grid()
plt.show()
```

9. Menampilkan tabel dataframe berdasarkan responden

```python
# Menghitung rataan opini/kenyataan responden (XR)
XR = np.array(d_opini).mean(axis = 1)
```

```python
# Menghitung rataan harapan responden (YR)
YR = np.array(d_harapan).mean(axis = 1)
```

```python
# Menghitung bobot responden (WR) dinyatakan dalam persen(%)
WR = (np.array(YR) / np.array(YR).sum()) * 100
```

```python
# Menghitung kenyataan bobot responden (WR * XR)
KBR = np.array(WR) * np.array(XR)
```

```python
# Membuat dataframe responden (DFR)
data_responden = {'Kenyataan (X)' : [x for x in XR],
        'Harapan (Y)' : [y for y in YR],
        'Bobot (W)' : [w for w in WR],
        'W x X' : [k for k in KBR]}
idx = [i for i in range(1, len(XR) + 1)]
DFR = pd.DataFrame(data_responden, index = [idx])
DFR
```

10. Menampilkan kategori CSI responden

```python
# Menghitung CSI responden
CSI_R = np.array(KBR).sum() / 5
CSI_R.round(2)
```

```python
# Kategori nilai CSI
if CSI_R >= 80.01:
    Kriteria = 'CSI : Sangat Puas'
elif CSI_R >= 65.01 and CSIP <= 80.00:
    Kriteria = 'CSI : Puas'
elif CSI_R >= 50.01 and CSIP <= 65.00:
    Kriteria = 'CSI : Cukup Puas'
elif CSI_R >= 35.01 and CSIP <= 50.00:
    Kriteria = 'CSi : Kurang Puas'
else:
    Kriteria = 'CSI : Tidak Puas'
Kriteria
```

11. Membuat grafik IPA untuk responden

```python
# Judul dan keterangan grafik
plt.title('Indicator Performance Analysis (IPA) Responden')
plt.xlabel('Kenyataan')
plt.ylabel('Harapan')

# Isi grafik (titik dan garis) serta keterangannya (legenda)
plt.scatter(XR, YR, color = 'grey', label = 'Data')
plt.plot([XR.mean(),XR.mean()],[YR.min()-0.1,YR.max()+0.1], color = 'red', label = 'Rataan Kenyataan')
plt.plot([XR.min()-0.2,XR.max()+0.2],[YR.mean(),YR.mean()], color = 'blue', label = 'Rataan Harapan')
plt.legend()
plt.grid()
plt.show()
```
